```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from auto_round import AutoRoundConfig
import os
from accelerate import Accelerator

os.environ['PYTORCH_CUDA_ALLOC_CONF'] = 'expandable_segments:True'
backend = "auto"  ##cpu, hpu, cuda
quantization_config = AutoRoundConfig(
    backend=backend
)
quantized_model_path = "kaitchup/Llama-3.3-70B-Instruct-AutoRoundGPTQ-2bit"

# إنشاء كائن Accelerator
accelerator = Accelerator(gradient_accumulation_steps=8, low_cpu_mem_usage=True)

# تحميل النموذج مع Accelerator
model = AutoModelForCausalLM.from_pretrained(
    quantized_model_path,
    device_map=backend.split(':')[0],
    quantization_config=quantization_config,
)
model = accelerator.prepare(model) # تجهيز النموذج للعمل مع Accelerator


tokenizer = AutoTokenizer.from_pretrained(quantized_model_path)
text = "There is a girl who likes adventure,"
inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512).to(model.device)  # Adjust max_length if needed

# تشغيل التوليد مع low_gpu_mem_usage
with accelerator.device_placement(): # لإدارة توزيع الأجهزة بشكل صحيح
    outputs = accelerator.unwrap_model(model).generate(
        **inputs, max_new_tokens=5, num_beams=1, low_gpu_mem_usage=True
    )

print(tokenizer.decode(outputs[0])) # Reduce num_beams if necessary
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from auto_round import AutoRoundConfig
import os
import torch

# تنظيف ذاكرة GPU
torch.cuda.empty_cache()
torch.cuda.reset_peak_memory_stats()

# ضبط إعدادات تخصيص الذاكرة في PyTorch
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"

# تحديد إعدادات التحميل
backend = "auto"  # سيتم توزيع النموذج تلقائيًا بين GPU و CPU
quantization_config = AutoRoundConfig(
    backend=backend
)

# تحديد المسار إلى النموذج
quantized_model_path = "kaitchup/Qwen2.5-14B-Instruct-gptqmodel-8bit"

# تعيين تخصيص الذاكرة لكل جهاز
max_memory = {
    0: "13GiB",  # تخصيص 6 جيجابايت على كودا 0
    1: "13GiB",  # تخصيص 6 جيجابايت على كودا 1
    "cpu": "30GiB"  # استخدام 30 جيجابايت على CPU
}

# تحميل النموذج مع التوزيع على الأجهزة المختلفة
model = AutoModelForCausalLM.from_pretrained(
    quantized_model_path,
    device_map="auto",
    max_memory=max_memory,
    quantization_config=quantization_config
)

# تحميل التوكنايزر
tokenizer = AutoTokenizer.from_pretrained(quantized_model_path)

# إدخال النص
text = "Who is Napoleon Bonaparte?"
inputs = tokenizer(text, return_tensors="pt").to(model.device)

# توليد النص
output = model.generate(**inputs, max_new_tokens=22)

# فك التشفير وطباعة الناتج
print(tokenizer.decode(output[0], skip_special_tokens=True))
```
