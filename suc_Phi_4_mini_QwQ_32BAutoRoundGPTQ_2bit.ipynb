```python
from transformers import AutoConfig, AutoModelForCausalLM, AutoTokenizer
from auto_round import AutoRoundConfig
import os

backend = "cuda"  # غيّره إلى "cpu" إذا كنت تستخدم المعالج فقط
local_model_path = r"/content/Phi-4-mini-instruct-AutoRoundGPTQ-2bit"

# تحميل التكوين من المجلد المحلي
config = AutoConfig.from_pretrained(local_model_path)

# التأكد من أن short_factor يحتوي على 64 قيمة
if "rope_scaling" in config.to_dict():
    short_factor = config.rope_scaling.get("short_factor", [])
    while len(short_factor) < 64:
        short_factor.append(1.0)  # إضافة القيم المفقودة

    config.rope_scaling["short_factor"] = short_factor

# تحميل النموذج من المجلد المحلي
quantization_config = AutoRoundConfig(backend=backend)
model = AutoModelForCausalLM.from_pretrained(
    local_model_path,
    config=config,
    device_map=backend.split(':')[0],
    quantization_config=quantization_config
)

# تحميل التوكنيزر من المجلد المحلي
tokenizer = AutoTokenizer.from_pretrained(local_model_path)

# تجربة النموذج
text = "Who is Einstein?"
inputs = tokenizer(text, return_tensors="pt").to(model.device)
print(tokenizer.decode(model.generate(**inputs, do_sample=True, max_new_tokens=55)[0]))
```

```python
import torch
text = "write python code for print hollow world"

# تحضير الإدخال للنموذج
inputs = tokenizer(text, return_tensors="pt").to(model.device)

# توليد الإخراج مع تحسين الإعدادات
with torch.no_grad():
    output = model.generate(
        **inputs,
        do_sample=True,         # تفعيل العشوائية
        max_new_tokens=55,      # الحد الأقصى لطول النص الناتج
        temperature=0.7,        # ضبط درجة العشوائية
        top_p=0.9,              # تقليل احتمالات الكلمات غير المناسبة
        eos_token_id=tokenizer.eos_token_id  # التوقف عند رمز النهاية
    )

# فك تشفير النص الناتج
print(tokenizer.decode(output[0], skip_special_tokens=True))
```
